```python
import matplotlib.pyplot as plt
import numpy as np

Y = [[480, 6490, 24030, 525180, 2056550, 54245130, 217916060],
     [3000, 15420, 32120, 178010, 386160, 2881870, 6229990]]
X = [10, 50, 100, 500, 1000, 5000, 10000]


plt.plot(X, Y[0], marker='o', markersize=7, label='Insertion sort')
plt.plot(X, Y[1], marker='o', markersize=7, label='Merge sort')

plt.legend(fontsize=14)
plt.xlabel('Array size') #Подпись для оси х
plt.ylabel('Time (nanosecond)') #Подпись для оси y
plt.title('Merge and Insertion sorts') #Название
#plt.semilogy()
plt.show()
```
